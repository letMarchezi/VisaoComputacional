# Processamento de textos com tranformers

Modelo bert minimizado com diferença entre palavras com caixa alta e baixa

```python
from transformers import pipeline
pipe = pipeline(model='distilbert/distilbert-base-cased', task='feature-extraction', 
                return_tensors=True)
```

```python
text = 'Some sentence to test the model'
y = pipe(text)
```

```python
y[0].shape
```

**Tokenizador**

Codificação do texto em representação numérica

**Modelo**

Modelo baseado em encoders (bert) para realização de uma tarefa de ML

```python
tokenizer = pipe.tokenizer
model = pipe.model

print(tokenizer)
print(model)
```

Em vez de mapear todas as palavras existentes, os tokens podem representar subpalavras.

Certos tokens, como '##ko' e '##herence' representam subwords, em que '##' representa que este token é uma continuação do token anterior.

```python
vocab = tokenizer.vocab
vocab
```

```python
text = 'Some sentence to test the model. Você conhece o modelo?\n\nNão'
y = pipe(text)
y
```

```python
text.encode()
```

A tokenização é realizada fazendo a codificação de cada token à um índice do vocabulário

```python
tokens = tokenizer.encode(text)
tokens
```

Tokens especiais como [CLS] e [SEP] são adicionados.

0 -> [PAD]: padding para completar o tamanho da sentença

100 -> [UNK]: token que não está presente no vocabulário

101 -> [CLS]: inicio de sentença

102 -> [SEP]: separador

103 -> [MASK]: máscara usada no treinamento

```python
tokenizer(text).tokens()
```

```python
texts = [text, 'Another sentence']
tokens = tokenizer(texts, return_tensors='pt', padding=True)
tokens
```

## Embeddings

Os ids dos tokens são transformados em vetores de valores. O número de linhas é o tamanho do vocabulário

```python
embedding = model.embeddings(tokens['input_ids'])

embedding.shape
```

O modelo gera duas matrizes, uma para cada sentença. Cada sentença possui 8 tokens. Para cada token, são gerados 768 atributos.

```python
res = model(**tokens)
res[0].shape
```

```python
text_features = res[0][:,0]
text_features.shape
```

## Codificador de texto

```python
from torch import nn
from transformers import pipeline
```

```python
class TextEncoder(nn.Module):
    
    def __init__(self):
        
        
        pipe = pipeline(model='distilbert/distilbert-base-cased', task='feature-extraction', 
                return_tensors=True)
        tokenizer = pipe.tokenizer
        model = pipe.model
        
        self.tokenizer = tokenizer
        self.model = model
        self.cls_token_id = 0
        self.feature_dim = 768
        
    def forward(self, text):
        
        tokens = self.tokenizer(text, (list, tuple))
        
```
